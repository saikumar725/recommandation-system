```python
import Recommenders
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
triplet=pd.read_csv('C:/Users/every/Downloads/innomatics/project/data/kaggle/kaggle_visible_evaluation_triplets.txt',sep='\t',names=['user_id','song_id','freq'])
```

```python
unique=pd.read_csv('C:/Users/every/Downloads/innomatics/project/data/kaggle/unique_tracks.txt',sep='<SEP>',names= ['track', 'song_id', 'title' ,'artist_name'])
```

```python
unique.head()
```

```python
song_df = pd.merge(triplet, unique, on = 'song_id')
```

```python
song_df
```

```python
song_df['song'] = song_df['title']+' - '+song_df['artist_name']
song_df.head()
```

```python
song_df = song_df.head(50000)
```

```python
song_grouped = song_df.groupby(['song']).agg({'freq':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['freq'].sum()
song_grouped['percentage'] = (song_grouped['freq'] / grouped_sum ) * 100
song_grouped.sort_values(['freq', 'song'], ascending=[0,1])
```
