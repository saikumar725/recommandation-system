```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
triplet=pd.read_csv('C:/Users/every/Downloads/innomatics/project/data/kaggle/kaggle_visible_evaluation_triplets.txt',sep='\t',names=['user_id','song_id','freq'])
```

```python
track=pd.read_csv('C:/Users/every/Downloads/innomatics/project/data/kaggle/unique_tracks.txt',sep='<SEP>',names= ['track', 'song_id', 'title' ,'artist_name'])
```

```python
song_df = pd.merge(triplet, track.drop_duplicates(['song_id']), on = 'song_id')
```

```python
song_df['song'] = song_df['title']+' - '+song_df['artist_name']
song_df.head()
```

```python
song_df=song_df.head(1000000)
songmat = song_df.pivot_table(index='user_id',columns='song',values='freq').fillna(0)
songmat.head(5)
```

```python
song_user_listen = songmat['Coldplay - Yellow']
similar=songmat.corrwith(song_user_listen )
df1=pd.DataFrame(similar,columns=['Correlation'])
df1.sort_values('Correlation',ascending=False).head(10)
```
